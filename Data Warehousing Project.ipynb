```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
columns_to_drop = [
    'Datasource', 'Topic', 'Data_Value_Unit', 'Data_Value_Type', 
    'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Total', 
    'ClassID', 'TopicID', 'QuestionID', 'DataValueTypeID', 
    'StratificationCategoryId1', 'StratificationID1', 'Education'
]

# Drop the specified columns
data_cleaned = df.drop(columns=columns_to_drop)
```

```python
data_cleaned.info()
```

```python
data_cleaned.head()
```

```python
def fill_age(row):
    if row['StratificationCategory1'] == 'Education':
        return np.random.randint(18, 26)  # Random age between 18 and 25
    elif row['StratificationCategory1'] == 'Income':
        return np.random.randint(20, 51)  # Random age between 20 and 50
    else:
        return np.random.randint(30, 61)  # Random age between 30 and 60 for other categories

# Apply the function to fill null ages based on the conditions
data_cleaned['Age(years)'] = data_cleaned.apply(lambda row: fill_age(row) if pd.isnull(row['Age(years)']) else row['Age(years)'], axis=1)
```

```python
null_gender_indices = data_cleaned[data_cleaned['Gender'].isnull()].index
data_cleaned.loc[null_gender_indices, 'Gender'] = np.random.choice(['Male', 'Female'], size=len(null_gender_indices))
```

```python
data_cleaned = data_cleaned.dropna(subset=['Income'])
```

```python
data_cleaned.info()
```

```python
data_cleaned = data_cleaned.drop(columns=['Race/Ethnicity'])
```

```python
data_cleaned = data_cleaned.dropna(subset=['GeoLocation'])
```

```python
data_cleaned = data_cleaned.dropna(subset=['Data_Value', 'Data_Value_Alt', 'Low_Confidence_Limit', 'Sample_Size'])
```

```python
data_cleaned.info()
```

```python
data = data_cleaned
```

```python
# Export the DataFrame to a CSV file
data.to_csv('data.csv', index=False)
```
